> Competencia: 1
>
> Ámbito conceptual: Análisis de oportunidades y riesgos en el emprendimiento. Evaluación de recursos y toma de decisiones en el emprendimiento.
>
> Objetivo: Reconocer y emplear con acierto los indicadores del estado financiero basico de negocios de emprendimiento.

### Segundo sello: para realizar en el cuaderno

1. Del primer juego escribe el concepto que mas te llamo la atencion y explica por que.

2. Escribe tres frases donde emplees al menos dos palabras de la sopa de letras, por ejemplo:

> las ___finanzas___ son indicadores para saber si un emprendimiento es ___exitoso___.

3. Selecciona tres o cuatro palabras de la sopa de letras y establece como se relacionan entre ellas por ejemplo:

> Presupuesto → Gastos → Costos
>
> Explicación: _Si establecemos correctamente el presupuesto podemos cubrir adecuadamente los gastos y los costos._
