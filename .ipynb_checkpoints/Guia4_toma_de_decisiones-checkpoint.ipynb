## Trabajo en Excel

1. Copie la siguiente tabla en Excel

|   Producto        |   Precio  |   Costo  | Unidades compradas |   Unidades vendidas  |   Ingreso  |   Costo total  |   Ganancia  |   Ganancia por unidad  |
|-------------------|-----------|----------|--------------------|----------------------|------------|----------------|-------------|------------------------|
|   Stickers        | 250       | 130      |                    |                      |            |                |             |                        |
|   Dulces          | 230       | 210      |                    |                      |            |                |             |                        |
|   Cuadernos       | 4500      | 3800     |                    |                      |            |                |             |                        |
|   Camisetas       | 15000     | 12500    |                    |                      |            |                |             |                        |
|   Pulseras        | 3600      | 2200     |                    |                      |            |                |             |                        |

|   Fundas celular  | 50000     | 43000    |                    |                      |            |                |             |                        |
|   Marcadores      | 6000      | 5200     |                    |                      |            |                |             |                        |

2. Complete con sus propios valores la columna `Unidades compradas`, escriba numeros realistas minimo 100 maximo 500
3. Complete con sus propios valores la columna `Unidades vendidas`, los valores deben ser inferiores a los escritos en unidades compradas, por ejemplo, _si usted como dueño del negocio compró 300 dulces no es posible que venda 400_ 🤦🏻‍♂️.
4. Emplee la formula necesaria para calcular el `ingreso`, es decir la cantidad de dinero que recibió por cada producto.
5. Emplee la formula necesaria para calcular el `costo total`, es decir la cantidad de dinero que gastó para la compra de todas las unidades.
6. Calcule en `Ganancia` el dinero total que recibió de ganancia por vender todos los productos.
7. Calcule en `Ganancia por unidad` el dinero que le gana a cada unidad que vende.

## Trabajo en el cuaderno

Finalmente conteste las siguiente pregunta en su cuaderno, conteste como si fuera el dueño del negocio:

1. Cual el producto de mayor ganancia total
2. Cual es el producto de mayor ganancia por unidad pero no tiene la mayor ganancia total?
3. Cual producto vende mucho pero deja poca ganancia?
4. Si solo pudiera escoger dos productos para negociar con ellos cuáles escogeria y por qué?
5. Es mejor vender mucho o ganar mas por unidad, justifique su respuesta.
