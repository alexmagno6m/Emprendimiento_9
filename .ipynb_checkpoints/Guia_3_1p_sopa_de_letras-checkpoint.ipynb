# Refinando los conceptos de un emprendimiento tecnológico

Esta actividad tiene un valor de dos sellos

### Primer sello: resuelve la sopa de letras

A continuación encontrarás una sopa de letras con __20 palabras__ relacionadas con emprendimiento y tecnología, debes solucionarla a través de la página que se proporciona haciendo clic en el enlace que aparece a continuación:

[Enlace sopa de letras](https://es.educaplay.com/recursos-educativos/28332637-emprendimiento_y_tecnologia_sopa_de_letras.html)

> Una vez la completes no cierres la ventana llama al profesor y escribe tu puntaje junto al sello

### Segundo sello: resuelve los siguientes puntos

1. Construye __tres frases__ donde uses en cada una al menos __dos palabras__ de la sopa de letras y subráyalas. Por ejemplo:

> _La tecnología permite crear aplicaciones que ayudan a los clientes_

2. Selecciona tres o cuatro palabras y explica como se relacionan entre ellas, por ejemplo:

> Idea → Problema → Solución → Clientes

### Obtención de la nota de los sellos

Finalmente escribe en tu cuaderno:

> Competencia: 1
>
> Ámbito conceptual: El emprendimiento tecnológico : concepto, características y aplicaciones en la vida cotidiana.
>
> Objetivo: Emplear asertivamente los conceptos relacionados en emprendimientos, estableciendo relaciones entre ellos apoyándose en una sopa de letras.
