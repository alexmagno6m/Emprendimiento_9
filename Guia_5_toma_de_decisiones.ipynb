## Operaciones financieras básicas

Como emprendedor y dueño del negocio se requiere poder calcular con precision los datos financieros necesarios para poder tomar decisiones:

> ### Ventas totales
> Corresponde a la multiplicacion entre el precio de venta por las unidades vendidas:
>
> ```Ventas totales = precio de venta unitario x unidades vendidas```

> ### Costos totales
> Corresponde a la multiplicación de costo unitario por unidades vendidas:
> 
> ```Costos totales = costo unitario x unidades vendidas```

> ### Ganancia por unidad
> Corresponde a la diferencia entre el precio de venta por unidad y el costo por unidad
>
> ```Ganancia por unidad = precio de venta unitario - costo unitario```

> ### Ganancias totales
> Corresponde a la diferencia entre ventas totales y costos totales:
>
> ```Ganancias totales = ventas totales - costos totales
