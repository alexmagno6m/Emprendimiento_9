# Introducción a la programación por bloques

Scratch es una herramienta visual que permite crear historias, animaciones, juegos y experiencias interactivas de forma sencilla. A través de bloques gráficos que se unen como piezas, los usuarios pueden construir proyectos donde personajes y objetos reaccionan a diferentes acciones, como presionar teclas, hacer clic o interactuar con otros elementos.

Más que aprender programación tradicional, Scratch permite desarrollar el pensamiento creativo y la capacidad de diseñar soluciones digitales. Mediante esta herramienta es posible comprender cómo funcionan muchas aplicaciones y sistemas interactivos que utilizamos diariamente, explorando conceptos como secuencia de acciones, eventos, decisiones y manejo de información de manera práctica y visual.

En el contexto del emprendimiento tecnológico, Scratch puede utilizarse para representar ideas, simular soluciones y construir prototipos interactivos que ayuden a comunicar propuestas de manera innovadora.

### Actividad

Una vez el profesor realice la introducción al tema y muestre un ejemplo, tendras que desarrollar diferentes retos.

1. En el cuaderno escribe una idea de emprendimiento
2. Una vez la hayas definido en Scratch, logra las siguientes acciones:

> - El personaje debe hablar de tu idea de emprendimiento
> - Se debe mover
> - Cambiar de color
> - Reproducir un sonido
> - Haz que pase algo al presionar una tecla

3. Contesta en tu cuaderno:

> - Que fue lo mas dificil de lograr
> - Que descrubriste explorando?
> - Que acción hizo que tu personaje reaccionara?

4. Finalmente, escribe los propositos de la clase:

> __Competencia__: 2
>
> __Ámbito conceptual__: Uso de herramientas digitales para la ideación, organización, análisis y la comunicación del emprendimiento
>
> __Objetivo__: cumplir con los retos propuestos empleando bloques basicos de una exploración autónoma se Scratch
